The data set contains six important features continuously monitored to evaluate the engine's status as ‘good’ or ‘bad’. These features are:
- **Engine rpm (revolutions per minute):**

A high rpm indicates the engine is operating at a higher speed than designed for prolonged periods, which can lead to overheating, excessive wear, and eventual failure.

A low rpm could signal a lack of power, issues with fuel delivery, or internal mechanical problems.

- **Lubrication oil pressure:**

Low lubrication oil pressure indicates insufficient lubrication, leading to increased friction, overheating, and engine damage.

A high lubrication oil pressure could signal a blockage in the oil delivery system, potentially causing seal or gasket failure.

- **Fuel pressure:**

High fuel pressure can cause poor engine performance and incomplete combustion, indicating fuel pump or filter issues.

A low fuel pressure may result in excessive fuel consumption, poor emissions, or damage to the fuel injectors.

- **Coolant pressure:**

Low coolant pressure indicates a potential leak in the cooling system or a coolant pump failure, risking engine overheating.

A high coolant pressure could be a sign of a blockage in the cooling system or a failing head gasket, which can also lead to overheating.

- **Lubrication oil temperature:**

High lubrication oil temperature suggests the oil is overheating, which can degrade its lubricating properties and lead to engine damage.

A low lubrication oil temperature may indicate it is not reaching its optimal operating temperature, potentially causing inadequate lubrication.

- **Coolant temperature:**

High coolant temperature signals overheating, which various issues, including a failed thermostat, coolant leak, or insufficient coolant flow can cause.

A low coolant temperature could suggest the engine is not reaching its optimal operating temperature, affecting performance and efficiency.

Issues with engines could lead to engine malfunctions, potential safety hazards, and downtime (e.g. delayed deliveries), resulting in the breakdown of a ship’s overall functionality, consequently impacting the business, such as affecting revenue via failure to deliver goods.
```python
# IQR-based Anomaly Detection for High and Low Outliers

import numpy as np

# We'll work with the cleaned data (df_cleaned) for IQR calculations
print("="*80)
print("IQR-BASED ANOMALY DETECTION - OUTLIER COUNTS")
print("="*80)

# Dictionary to store outlier information
outlier_info = {}

for feature in df_cleaned.columns:
    # Calculate Q1, Q3, and IQR
    Q1 = df_cleaned[feature].quantile(0.25)
    Q3 = df_cleaned[feature].quantile(0.75)
    IQR = Q3 - Q1

    # Define lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify low outliers (below lower bound)
    low_outliers = df_cleaned[df_cleaned[feature] < lower_bound]
    num_low_outliers = len(low_outliers)

    # Identify high outliers (above upper bound)
    high_outliers = df_cleaned[df_cleaned[feature] > upper_bound]
    num_high_outliers = len(high_outliers)

    # Store the information
    outlier_info[feature] = {
        'Q1': Q1,
        'Q3': Q3,
        'IQR': IQR,
        'lower_bound': lower_bound,
        'upper_bound': upper_bound,
        'num_low_outliers': num_low_outliers,
        'num_high_outliers': num_high_outliers,
        'low_outlier_indices': low_outliers.index.tolist(),
        'high_outlier_indices': high_outliers.index.tolist()
    }

# Print results
total_samples = len(df_cleaned)
print(f"\nTotal number of samples: {total_samples}\n")

for feature in df_cleaned.columns:
    info = outlier_info[feature]
    print(f"\n{feature}:")
    print(f"  Q1: {info['Q1']:.4f}, Q3: {info['Q3']:.4f}, IQR: {info['IQR']:.4f}")
    print(f"  Lower bound: {info['lower_bound']:.4f}, Upper bound: {info['upper_bound']:.4f}")
    print(f"  LOW outliers:  {info['num_low_outliers']:5d} ({info['num_low_outliers']/total_samples*100:.2f}%)")
    print(f"  HIGH outliers: {info['num_high_outliers']:5d} ({info['num_high_outliers']/total_samples*100:.2f}%)")

print("\n" + "="*80)
print("SUMMARY TABLE")
print("="*80)
print(f"{'Feature':<20} {'Low Outliers':>15} {'High Outliers':>15} {'Total':>10}")
print("-"*80)
for feature in df_cleaned.columns:
    info = outlier_info[feature]
    total_for_feature = info['num_low_outliers'] + info['num_high_outliers']
    print(f"{feature:<20} {info['num_low_outliers']:>15} {info['num_high_outliers']:>15} {total_for_feature:>10}")
print("="*80)
```

```python
# Calculate how many samples would be identified at 1%, 3%, and 5% thresholds

print("\n" + "="*80)
print("ANOMALY PERCENTAGES AT 1%, 3%, AND 5% THRESHOLDS")
print("="*80)

# Calculate the number of samples for each percentage threshold
threshold_1_percent = int(total_samples * 0.01)
threshold_3_percent = int(total_samples * 0.03)
threshold_5_percent = int(total_samples * 0.05)

print(f"\nTotal samples: {total_samples}")
print(f"1% threshold: {threshold_1_percent} samples")
print(f"3% threshold: {threshold_3_percent} samples")
print(f"5% threshold: {threshold_5_percent} samples")

print("\n" + "-"*80)
print("OUTLIER COUNTS FOR EACH FEATURE (HIGH vs LOW)")
print("-"*80)
print(f"{'Feature':<25} {'Low':>10} {'High':>10} {'Total':>10} {'% of Data':>12}")
print("-"*80)

for feature in df_cleaned.columns:
    info = outlier_info[feature]
    total = info['num_low_outliers'] + info['num_high_outliers']
    pct = (total / total_samples) * 100
    print(f"{feature:<25} {info['num_low_outliers']:>10} {info['num_high_outliers']:>10} {total:>10} {pct:>11.2f}%")

print("="*80)
```

```python
# Create Binary IQR Outlier Flags for All 12 Dimensions

import pandas as pd
import numpy as np

# Create a DataFrame to hold binary flags for each of the 12 dimensions
# We'll use the cleaned data as our base
df_iqr_flags = pd.DataFrame(index=df_cleaned.index)

# Map of dimension names to (feature, direction)
dimensions_map = [
    ('Engine RPM - HIGH', 'Engine rpm', 'high'),
    ('Engine RPM - LOW', 'Engine rpm', 'low'),
    ('Lub oil pressure - HIGH', 'Lub oil pressure', 'high'),
    ('Lub oil pressure - LOW', 'Lub oil pressure', 'low'),
    ('Fuel pressure - HIGH', 'Fuel pressure', 'high'),
    ('Fuel pressure - LOW', 'Fuel pressure', 'low'),
    ('Coolant pressure - HIGH', 'Coolant pressure', 'high'),
    ('Coolant pressure - LOW', 'Coolant pressure', 'low'),
    ('Lub oil temp - HIGH', 'lub oil temp', 'high'),
    ('Lub oil temp - LOW', 'lub oil temp', 'low'),
    ('Coolant temp - HIGH', 'Coolant temp', 'high'),
    ('Coolant temp - LOW', 'Coolant temp', 'low'),
]

# Create binary flags for each dimension
for dim_name, feature, direction in dimensions_map:
    info = outlier_info[feature]

    if direction == 'high':
        # Flag samples above upper bound
        df_iqr_flags[dim_name] = (df_cleaned[feature] > info['upper_bound']).astype(int)
    else:  # low
        # Flag samples below lower bound
        df_iqr_flags[dim_name] = (df_cleaned[feature] < info['lower_bound']).astype(int)

# Calculate total number of outlier dimensions per sample
df_iqr_flags['Total_IQR_Flags'] = df_iqr_flags.sum(axis=1)

print("="*100)
print("IQR BINARY FLAGS - 12 DIMENSIONS")
print("="*100)
print("\nFirst 10 rows of IQR flags DataFrame:")
print(df_iqr_flags.head(10))

print("\n" + "="*100)
print("DISTRIBUTION OF TOTAL IQR FLAGS PER SAMPLE")
print("="*100)
flag_distribution = df_iqr_flags['Total_IQR_Flags'].value_counts().sort_index()
print(f"\n{'Num Flags':<15} {'Count':>10} {'Percent':>10} {'Cumulative %':>15}")
print("-"*100)
cumulative = 0
for num_flags, count in flag_distribution.items():
    pct = (count / len(df_iqr_flags)) * 100
    cumulative += pct
    print(f"{num_flags:<15} {count:>10} {pct:>9.2f}% {cumulative:>14.2f}%")

print("\n" + "="*100)
print("SUMMARY STATISTICS")
print("="*100)
print(f"Samples with 0 flags (no outliers): {(df_iqr_flags['Total_IQR_Flags'] == 0).sum()} ({(df_iqr_flags['Total_IQR_Flags'] == 0).sum()/len(df_iqr_flags)*100:.2f}%)")
print(f"Samples with 1+ flags (any outlier): {(df_iqr_flags['Total_IQR_Flags'] >= 1).sum()} ({(df_iqr_flags['Total_IQR_Flags'] >= 1).sum()/len(df_iqr_flags)*100:.2f}%)")
print(f"Samples with 2+ flags (multi-feature): {(df_iqr_flags['Total_IQR_Flags'] >= 2).sum()} ({(df_iqr_flags['Total_IQR_Flags'] >= 2).sum()/len(df_iqr_flags)*100:.2f}%)")
print(f"Samples with 3+ flags (triplet+): {(df_iqr_flags['Total_IQR_Flags'] >= 3).sum()} ({(df_iqr_flags['Total_IQR_Flags'] >= 3).sum()/len(df_iqr_flags)*100:.2f}%)")
print(f"Maximum flags in a single sample: {df_iqr_flags['Total_IQR_Flags'].max()}")
print("="*100)
```
